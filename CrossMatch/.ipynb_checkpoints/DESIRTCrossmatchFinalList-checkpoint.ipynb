```python
# Here I'm just using this to import the relevant packages, not actually using any of the functions in this file
#from crossmatch_fns import *
import pandas as pd
import numpy as np
from astropy.coordinates import SkyCoord
from astropy.wcs import WCS
from astropy import units as u
import matplotlib.pyplot as plt
```

```python
'''
Get the data, make it nice
'''

#Overall path
semesters_folder_path = '/hildafs/projects/phy220048p/leihu/AstroWork/DESIRT/IMPROC/trigger/MESS/FinalList'

# Path to each summary file
path_to_21A_summ = f'{semesters_folder_path}/LYSelection-21A.CandidateSummary_RefinedDLR.csv'
path_to_22A_summ = f'{semesters_folder_path}/LYSelection-22A.CandidateSummary_RefinedDLR.csv'
path_to_22B_summ = f'{semesters_folder_path}/LYSelection-22B.CandidateSummary_RefinedDLR.csv'
path_to_23A_summ = f'{semesters_folder_path}/LYSelection-23A.CandidateSummary_RefinedDLR.csv'
path_to_23B_summ = f'{semesters_folder_path}/LYSelection-23B.CandidateSummary_RefinedDLR.csv'

# Columns I need right now
usecols = ['objid', 'ra_obj', 'dec_obj']

# Create dataframes with just the columns we want
df_21A = pd.read_csv(path_to_21A_summ, usecols=usecols).dropna()
df_22A = pd.read_csv(path_to_22A_summ, usecols=usecols).dropna()
df_22B = pd.read_csv(path_to_22B_summ, usecols=usecols).dropna()
df_23A = pd.read_csv(path_to_23A_summ, usecols=usecols).dropna()
df_23B = pd.read_csv(path_to_23B_summ, usecols=usecols).dropna()

# Make a list for convenience later
dfs = [df_21A, df_22A, df_22B, df_23A, df_23B]

# Create the 'skycoord_obj' column from 'ra_obj' and 'dec_obj'
# This column is just a concatenation of ra_obj and dec_obj as strings
# Ex: If ra_obj = 100 and dec_obj = 50.3, skycoord_obj = '100 50.3'
for df in dfs:
    df['skycoord_obj'] = df['ra_obj'].astype(str).str.cat(df['dec_obj'].astype(str), sep=' ')

# Sort each df by coordinate string and allow for reindexing
# Every data frame is now sorted by the skycoord_obj column, essentially sorted by coordinate
[df.sort_values('skycoord_obj', ignore_index=True, inplace=True) for df in dfs]

# Add a string 'sem' column to each dataframe that specifies the semester it is from
# This is used to identify which semester each object is coming from when we start mixing them
dfs_names = [(df_21A, '21A'), (df_22A, '22A'), (df_22B, '22B'), (df_23A, '23A'), (df_23B, '23B')]
for df, name in dfs_names:
    df['sem'] = name
    
# Threshold for matching
sep_cutoff = 1 #arcsec
```

```python
'''
This crossmatch acknowledges that matches might not be consecutive when sorted by skycoord_obj
Ex: 
skycoord1 = '100.0 50.0'
skycoord2 = '100.0 300.0'
skycoord3 = '100.1 50.0'
These are sorted correctly, but skycoord1 and skycoord3 should be matched. I fix this just by searching farther, since
matches may occasionally be non-consecutive but they will certainly be very close when sorted like this. So far actually I have
never found a non-consecutive match, I just know it's possible


At each iteration, it takes the 5 (set by search_depth) 'least' elements of each semester dataframe and puts them all into a new min_df.
Take the least element of this new dataframe and find its separation to all the others in min_df. Then take that least element and 
anything it matched with, and put them in a new row of crossmatch_df, and remove them from their respective dataframes.

NOTE: When you run this, it 'destroys' the semester dataframes as it goes, so if you want to rerun the crossmatch, you have to rerun the 
    cell above the get the original semester dataframes back
'''

search_depth = 5 # Sets how many objects we take from each semester at a time. Set this higher to be more confident that you aren't
                 # missing non-consecutive matches as in the example above. Crossmatch takes longer if this is increased

# For convenience later
name_to_df = {
    '21A': df_21A,
    '22A': df_22A,
    '22B': df_22B,
    '23A': df_23A,
    '23B': df_23B,
}

# Initialize but empty. It will build itself up as it goes
crossmatch_df = pd.DataFrame()

iter_count = 0 # Count number of iterations of the while loop
match_count = 0 # Count number of matches (If we find match two objects across semesters, count that as one match)

seps = [] # List of separations in arcseconds for each individual match, for histogram

# While there's at least one non-empty semester dataframe
while np.any([df.shape[0]>0 for df in dfs]):
    
    # Get the first search_depth rows of each dataframe, put them into a new dataframe, sort by skycoord_obj, reindex
    min_df = pd.concat([df.iloc[0:search_depth] for df in dfs if df.shape[0]>0]).sort_values('skycoord_obj', ignore_index=True)
    
    
    # List of indeces in min_df that we want to match together in crossmatch_df
    # Always going to take the least element at index 0, so we put it in there manually first
    # Any other matching index will be added to this
    use_indeces = [0]
    
    # Get the coord of the minimum element
    min_coord = min_df.loc[0,'skycoord_obj']
    min_skycoord = SkyCoord(min_coord, unit=u.deg)
    
    # Compare the least element to every other element in min_df
    for i in range(1, min_df.shape[0]-1):
        # Get the coord of the current element
        coord2 = min_df.loc[i,'skycoord_obj']
        skycoord2 = SkyCoord(coord2, unit=u.deg)
        
        # Find the seperation between min_coord and this new coord
        sep = min_skycoord.separation(skycoord2)
        
        # Check if they are a match
        if sep.arcsecond<sep_cutoff:
            
            ''' Print matches and separation for debugging
            name1 = min_df.loc[0,'objid']
            name2 = min_df.loc[i,'objid']
            if name1[0]!=name2[0]:
                print(f"matched {name1} to {name2} with separation {sep.arcsecond}\"")
            '''
            
            seps.append(sep.arcsecond)
            
            # Add this index to use_indeces
            use_indeces.append(i)
    
    # Check if we found a match
    if len(use_indeces)>1:
        # Found a match!
        match_count += 1
            
    # New dataframe with matched objects to add to crossmatch dataframe and remove from their respective initial dataframes
    add_df = min_df.loc[use_indeces]
    crossmatch_rows = crossmatch_df.shape[0] # Get num rows of crossmatch currently before adding stuff, so we can easily add new info in a new bottom row
    
    # Initialize lists to store the coordinates of a single object so we can average them if necessary
    ra_obj_list = [] 
    dec_obj_list = []
    
    # Iterate through the matched objects we want to add
    for i, row in add_df.iterrows():
        # Get the values we want from the row
        objid = row['objid']
        skycoord_obj = row['skycoord_obj']
        ra_obj = row['ra_obj']
        dec_obj = row['dec_obj']
        sem = row['sem']
        
        # Add the coordinates to their respective lists
        ra_obj_list.append(ra_obj)
        dec_obj_list.append(dec_obj)
        
        # Add the objid to crossmatch_df (at the bottom)
        crossmatch_df.loc[crossmatch_rows, sem] = objid
        
        # Get the dataframe from which we got this object
        sem_df = name_to_df[sem]
        # Get the 'small' version that only contains the data we just looked at (speeds up search in the next line)
        sem_df_small = sem_df.iloc[0:search_depth]
        # Find the index of the objid we want to drop and drop it
        sem_df.drop(sem_df_small[sem_df_small['objid'] == objid].index, inplace=True)

    # Add average ra_obj and dec_obj to crossmatch_df
    crossmatch_df.loc[crossmatch_rows, 'ra_obj'] = np.mean(ra_obj_list)
    crossmatch_df.loc[crossmatch_rows, 'dec_obj'] = np.mean(dec_obj_list)
    
        
    # Print progress every 500 iterations
    if iter_count%500==0:
        rem_objs = np.sum([df.shape[0] for df in dfs])
        print(f"{iter_count} iterations complete - {rem_objs} objects left to be matched - {match_count} matches found")
    iter_count += 1
    
# Print final status
rem_objs = np.sum([df.shape[0] for df in dfs])
print(f"FINAL: {iter_count} iterations complete - {rem_objs} objects left to be matched - {match_count} matches found")

# Reorder columns for convenience
crossmatch_df = crossmatch_df[['21A', '22A', '22B', '23A', '23B', 'ra_obj', 'dec_obj']]
```

```python
plt.hist(seps, bins=50, range=(0,sep_cutoff))
```

```python
crossmatch_df.to_csv('DESIRT_one_arcsecond_crossmatch_final_list.csv', index=False)
```
